```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# Xem biểu đồ trong Jupyter Notebook
%matplotlib inline 
print("Setup Complete")
```

```python
# Initialize a Figure
fig = plt.figure()

# Add Axes to the Figure
fig.add_axes([0,0,1,1])
```

```python
# Line graph

plt.plot([1,2],[3,4]) # vẽ đường thằng từ hai điểm (x,y) = (1,3); (2,4)
plt.title("My First Plot")
plt.xlabel("My x-axis")
plt.ylabel("My y-axis")
plt.legend(["Line 1"])
plt.show()
```

```python
# Bar graph

#create data for plotting
y_values = [5,6,3,7,2]
x_values = ["A", "B", "C", "D", "E"]
plt.bar(x_values, y_values, color = "green")
plt.show()
```

```python
#create data for plotting

x_values = [5,6,3,7,2]
y_val = ["A", "B", "C", "D", "E"]
# Adding an "h" after bar will flip the graph
plt.barh(y_val,x_values, color ="yellowgreen")
plt.show()
```

```python
# Scatter plot

#create data for plotting
x_values = [0,1,2,3,4,5]
squares = [0,1,4,9,16,25]
plt.scatter(x_values,squares, s=10, color = "pink")
plt.show()
```

```python
# Histogram graph

#generate data
x = [2,1,6,4,2,4,8,9,4,2,4,10,6,4,5,7,7,3,2,7,5,3,5,9,2,1]
#plot for a histogram
plt.hist(x, bins = 10, color='green', alpha=1)
plt.show()
```

Đây là biểu đồ tần suất các giá trị.

Bins: là đối số thể hiện số cột bạn muốn.

Alpha: là đối số thể hiện mức độ rõ ràng của dữ liệu.

Trên đây là khởi đầu nhẹ nhàng cho mọi người làm quen với biểu đồ. Chi tiết hơn, mọi người xem tiếp ở phần sau nhé :))
